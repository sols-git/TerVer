Похоже, особых совпадений нет. Или есть? Нельзя сказать точно, пока не посчитаем корреляцию. Расчёт корреляции — это самый простой способ определить степень линейной связи двух числовых рядов (в том числе и временных).

```python
# для расчёта корреляции необходимо удалить пропущенные значения

corr = df.dropna().hc.corr(df.dropna().ore1)
print("Correlation is {:.3}".format(corr))
```

```python
for i in range(1,len(df.columns)-1):
    col_x= df.columns[1]
    col_y= df.columns[i+1]
    x = df[col_x]
    y = df[col_y]
    print(f'Kpirsn({col_x}, {col_y}) - {np.corrcoef(x, y)[0,1]}')
```

Корреляция более 0.66 обычно считается сильной, а у нас есть 0.95. Можно ли этим удовлетвориться? Нет!

Применяя корреляция к значениями временного ряда вы будете стабильно получать её высокое значение. Это происходит потому что исходные значения временных рядов как правило нестационарны. Решением может быть использование для корреляции не исходные данные о ценах, а процентное изменение этих цен по сравнению с предыдущим шагом.

```python
print('Корреляция процентного изменения между:\n')
for col in range(1,len(df.columns)-1):
    pct_corr = df.dropna().hc.pct_change().corr(df[df.dropna().columns[col+1]].pct_change())
    print(f'{col}. HC и {df.columns[col+1]} - {pct_corr}')
```

И тут кореляция уже не такая сильная, а наоборот очень маленькая. Надо отметить на самую большую из этого перечня корреляцию с Шанхайским фьючерсом горячекатной стали HC и арматуры RB. Это вполне закономерно, так как арматура производится из стали и их цены в абсолютном значении торгуются очень близко. Но их слабая корреляция процентного изменения стала для меня сюрпризом 

Если мы построим график, в котором на одной оси отложим цену на сталь, а на другом фьючерс на арматуру, то как раз можно увидеть их похожесть при определении простой корреляции, во втором случае, корреляции процентного изменения, график более равномерный но корреляция стала слабая. Возможно это сваязано с различной волатильностью данных фьючерсов. У арматуры волатильность несколько больше из за того, что для пее производства не требются большие производственные мощности.

```python
plt.subplot(1, 2, 1)
plt.scatter(
    df.dropna().hc,
    df.dropna().rb);
plt.title('Correlation')

plt.subplot(1, 2, 2)
plt.scatter(
    df.dropna().hc.pct_change(),
    df.dropna().rb.pct_change());
plt.title('Percent change correlation');
```

Степень линейной связи двух числовых рядов ещё можно устаноить при помощи регрессионной модели, из statsmodels, функции OLS. В качестве зависимой переменной укажем фьючерс на арматуру RB, а на месте независимой — цену на Шанхайский фьючерс горячекатной рулонной стали и свободный коэффициент, который заранее добавим фукнцией add_constant:

```python
def research_regression(df, dependent_series, independen_series):
    # Add a constant to the DataFrame x
    reg_def = sm.add_constant(df.dropna())

    # Fit the regression of y on x
    result = sm.OLS(reg_def[independen_series], reg_def[['const', dependent_series]]).fit()

    # Print out the results
    result
    print(result.summary())
```

```python
research_regression(df, 'hc', 'rb')
```

Регрессионная модель показывает, что два временныз ряда значимо связаны, причём цена на арматуру объясняет 91.8% дисперсии индекса (R-squared: 0.918). Исходя из этого мы можем сказать, что связь между Шанхакйским фьючерсом горячекатной стали и арматуры определенно есть и это можно использовать в дальшейшем для пронозировании цен на сталь применяя стоимость арматуры в качестве одного из регрессоров.

### Для самомпроверки пересчитаем уровни значимости для всех временных рядов

```python
def sum_of_squares(samples: np.ndarray) -> float:
    return ((samples - samples.mean()) ** 2).sum()
```

```python
def f_chose(x, y):
    factor = pd.DataFrame(x).columns
    ones = np.ones((x.shape[0], 1))
    x = np.column_stack((ones, x))
    XTX = x.T.dot(x)
    det = np.linalg.det(XTX)
    XTX_inv = np.linalg.inv(XTX)
    b = XTX_inv.dot(x.T).dot(y)
    z = x.dot(b)
    e = y - z
    ae = e.mean()
#     print('average error = ', float(ae))
    R = 1 - sum_of_squares(e) / sum_of_squares(y)
#     print('R =', float(R))
    z.var(), y.var()
    n = x.shape[0]
    m = 1
    k1 = m
    k2 = n - m - 1
    k1, k2
    alpha = 0.05
    t = stats.f.ppf(1 - alpha, k1, k2)
#     print('quantile =', t)
    F = (R / k1) / ((1 - R) / k2)
#     print('statistic=', float(F))
    significancy = ''
    if float(F) > float(t):
        significancy = 'уравнение регрессии признаётся значимым'
    else:
        significancy = 'уравнение регрессии НЕ признаётся значимым'
       
    info = {'factor': factor, 'determ': det, 'average error': float(ae), 'R': float(R), 'statistic':float(F), 'significancy': significancy}
    return info
```

```python
y = df['hc']
X = df
X.drop(['hc'], axis='columns', inplace=True)
```
